```python
import pandas as pd 
import numpy as np 
```

```python
cc = pd.read_csv('/Users/macbookpro/Downloads/customer_churn_dataset.csv')
```

```python
cc.head()
```

```python
cc.tail()
```

```python
cc.shape
```

```python
cc.info()
```

```python
cc.isnull().sum()
```

```python
cc.Age.value_counts()
```

```python
cc.Age = cc.Age.fillna(cc.Age.mean())
```

```python
cc.Gender = cc.Gender.fillna('Male')
```

```python
cc.Tenure.value_counts()
```

```python
cc.Tenure = cc.Tenure.fillna(cc.Tenure.mean())
```

```python
cc.columns = cc.columns.str.replace(' ', '_')
```

```python
cc.Usage_Frequency = cc.Usage_Frequency.fillna(cc.Usage_Frequency.mean())
```

```python
cc.Support_Calls = cc.Support_Calls.fillna(0.0)
```

```python
cc.Payment_Delay = cc.Payment_Delay.fillna(cc.Payment_Delay.mean())
```

```python
cc.Subscription_Type = cc.Subscription_Type.fillna('Standard')
```

```python
cc.Contract_Length = cc.Contract_Length.fillna('Annual')
```

```python
cc.Total_Spend = cc.Total_Spend.fillna(cc.Total_Spend.mean())
```

```python
cc.Last_Interaction = cc.Last_Interaction.fillna(cc.Last_Interaction.mean())
```

```python
cc.Churn = cc.Churn.fillna(1.0)
```

```python
cc.drop('CustomerID', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cc[cc.select_dtypes(include=['object']).columns] = cc[cc.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
cc_train, cc_test = train_test_split(cc, test_size=0.2)
```

```python
cc_train_x = cc_train.iloc[:, :-1]
cc_train_y = cc_train.iloc[:, -1]
```

```python
cc_test_x = cc_test.iloc[:, :-1]
cc_test_y = cc_test.iloc[:, -1]
```

```python
cc_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cc_lr = LogisticRegression()
```

```python
cc_lr.fit(cc_train_x, cc_train_y)
```

```python
pred = cc_lr.predict(cc_test_x)
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(cc_test_y, pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(cc_test_y, pred)*100
```

```python
from sklearn.metrics import classification_report
```

```python
print('\nClassification Report:')
print(classification_report(cc_test_y, pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cc_dectree = DecisionTreeClassifier()
```

```python
cc_dectree.fit(cc_train_x, cc_train_y)
```

```python
dt_pred = cc_dectree.predict(cc_test_x)
dt_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(cc_test_y, dt_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(cc_test_y, dt_pred)*100
```
